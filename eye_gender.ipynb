```python
#importing libraries

import numpy as np
import matplotlib.pyplot as  plt
import cv2
import tensorflow as tf
import os
```

```python
import pandas as pd

labels = pd.read_csv("eye_gender_data/Training_set.csv")   # loading the labels
file_paths = [[fname, 'eye_gender_data/train/' + fname] for fname in labels['filename']]
images = pd.DataFrame(file_paths, columns=['filename', 'filepaths'])
train_data = pd.merge(images, labels, how = 'inner', on = 'filename')

data = []     # initialize an empty numpy array
image_size = 100      # image size taken is 100 here. one can take other size too
for i in range(len(train_data)):
  
  img_array = cv2.imread(train_data['filepaths'][i], cv2.IMREAD_GRAYSCALE)   # converting the image to gray scale

  new_img_array = cv2.resize(img_array, (image_size, image_size))      # resizing the image array
  data.append(new_img_array)
```

```python
#convert to an array
data_arr = np.array(data)
```

```python
#reshaping the data
train_image_reshp =  np.stack([data_arr]*3, axis=-1)
```

```python
#normalizing 

train_image_norm = train_image_reshp/255
train_image_norm
```

```python
#let's addresss the labels. we'll encode it to have 1 and 0
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_labels = le.fit_transform(train_data[['label']])
```

```python
#model training
#let's try transfer learning using VGG_19

INPUT_SHAPE = (100, 100, 3)
vgg_layers = tf.keras.applications.vgg19.VGG19(weights='imagenet', include_top= False, input_shape=INPUT_SHAPE)

vgg_layers.summary()
```

```python
#let's make the layers trainable
for layer in vgg_layers.layers:
    layer.trainable = True

for layer in vgg_layers.layers:
    print(layer, layer.trainable )
```

```python
#lets attempt a cnn on the vgg model

# define sequential model
model = tf.keras.models.Sequential()

# Add the vgg convolutional base model
model.add(vgg_layers)

# add flatten layer
model.add(tf.keras.layers.Flatten())

# add dense layers with some dropout
model.add(tf.keras.layers.Dense(256, activation='relu'))
model.add(tf.keras.layers.Dropout(rate=0.3))
model.add(tf.keras.layers.Dense(256, activation='relu'))
model.add(tf.keras.layers.Dropout(rate=0.3))

# add output layer
model.add(tf.keras.layers.Dense(10, activation='softmax'))

# compile model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=2e-5), 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

# view model layers
model.summary()
```

```python
EPOCHS = 100

es_callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, 
                                               restore_best_weights=True,
                                               verbose=1)

history = model.fit(train_image_norm, train_labels,
                    batch_size=32,
                    callbacks=[es_callback], 
                    validation_split=0.1, epochs=EPOCHS,
                    verbose=1)
```

```python
#let's plot the learning curves
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

history_df = pd.DataFrame(history.history)
history_df[['loss', 'val_loss']].plot(kind='line', ax=ax[0])
history_df[['accuracy', 'val_accuracy']].plot(kind='line', ax=ax[1]);
```

```python
# saving the model
save_dir = "/results/"
model_name = 'eye_model.h5'
model.save(model_name)
model_path = save_dir + model_name
print('Saved trained model at %s ' % model_path)
```
